```python
import matplotlib.pyplot as plt
import math
from skimage.transform import rotate
from skimage.transform import EuclideanTransform
from skimage.transform import SimilarityTransform
from skimage.transform import warp
from skimage.transform import swirl
from skimage import io
```

```python
# Utility function

def show_comparison(original, transformed, transformed_name):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 4), sharex=True,
                                   sharey=True)
    ax1.imshow(original)
    ax1.set_title('Original')
    ax1.axis('off')
    ax2.imshow(transformed)
    ax2.set_title(transformed_name)
    ax2.axis('off')
    io.show()
```

```python
# Open image
im_org = io.imread('data/NusaPenida.png')
```

```python
# Image rotation
# Angle in degrees, counter-clockwise
rotation_angle = 10
rotated_img = rotate(im_org, rotation_angle)
show_comparison(im_org, rotated_img, "Rotated image")
```

```python
# Rotation with a different center
rot_center = [0, 0]
rotated_img = rotate(im_org, rotation_angle, center=rot_center)
show_comparison(im_org, rotated_img, "Rotated image")
```

```python
# Rotation with background fill
rotated_img = rotate(im_org, rotation_angle, mode="reflect")
rotated_img_2 = rotate(im_org, rotation_angle, mode="wrap")
show_comparison(im_org, rotated_img, "Rotated image with reflect mode")
show_comparison(im_org, rotated_img_2, "Rotated image with wrap mode")
```

```python
# Rotation with cval
rotated_img = rotate(im_org, rotation_angle, mode="constant", cval=255)
show_comparison(im_org, rotated_img, "Rotated image with constant mode")
```

```python
# Rotation with automatic resize
rotated_img = rotate(im_org, rotation_angle, resize=True)
rotated_img_2 = rotate(im_org, rotation_angle, resize=True, cval=255)
show_comparison(im_org, rotated_img, "Rotated image with resize")
show_comparison(im_org, rotated_img_2, "Rotated image with resize and cval")
```
