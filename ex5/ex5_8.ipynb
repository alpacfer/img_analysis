```python
from skimage.filters import prewitt_h
from skimage.filters import prewitt_v
from skimage.filters import prewitt
from skimage import io
import matplotlib.pyplot as plt
from skimage.filters import median, gaussian
import numpy as np
from skimage.filters import threshold_otsu
```

```python
# Read image
img = io.imread('data/ElbowCTSlice.png', as_gray=True)

# Show image
plt.figure()
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.title('Original image')
plt.show()
```

```python
# Filterint with Gaussian filter
img_gauss = gaussian(img, sigma=4)

# Show image
plt.figure()
plt.imshow(img_gauss, cmap='gray')
plt.axis('off')
plt.title('Gaussian filter')
plt.show()
```

```python
# Filterint with Median filter
size = 10
footprint = np.ones((size, size))
img_median = median(img, footprint=footprint)

# Show image
plt.figure()
plt.imshow(img_median, cmap='gray')
plt.axis('off')
plt.title('Median filter')
plt.show()
```

```python
# Apply Prewitt filter
img_prewitt = prewitt(img_median) # Use median filtered image

# Show image
plt.figure()
plt.imshow(img_prewitt, cmap='gray')
plt.axis('off')
plt.title('Prewitt filter')
plt.show()
```

```python
# Apply Ostus threshold
thresh = threshold_otsu(img_prewitt)
img_binary = img_prewitt > thresh

# Show image
plt.figure()
plt.imshow(img_binary, cmap='gray')
plt.axis('off')
plt.title('Binary image')
plt.show()
```

```python
# Visualization with terrain color map

min_val = np.min(img)
max_val = np.max(img)


plt.figure()
plt.imshow(img, vmin=min_val, vmax=max_val, cmap='terrain')
plt.axis('off')
plt.title('Original image')
```
