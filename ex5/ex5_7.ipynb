```python
from skimage.filters import prewitt_h
from skimage.filters import prewitt_v
from skimage.filters import prewitt
from skimage import io
import matplotlib.pyplot as plt
```

```python
# Read the image
img = io.imread('data/donald_1.png', as_gray=True)
```

```python
# Filter with PreWitt
prewitt_h_img = prewitt_h(img) # Horizontal
prewitt_v_img = prewitt_v(img) # Vertical

# Show the images
plt.figure(figsize=(15,15))
plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.subplot(1,3,2)
plt.imshow(prewitt_h_img, cmap='gray')
plt.title('Prewitt Horizontal')
plt.subplot(1,3,3)
plt.imshow(prewitt_v_img, cmap='gray')
plt.title('Prewitt Vertical')
plt.show()
```

```python
# Use full Prewitt filter
prewitt_img = prewitt(img)

# Show the images
plt.figure(figsize=(15,15))
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(prewitt_img, cmap='gray')
plt.title('Prewitt')
plt.show()
```
